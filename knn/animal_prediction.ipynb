```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("../datasets/animals.csv")
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
print(X)
```

```python
print(y)
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
print(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=1)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
predictions = knn.predict(X_test)
decoded_predictions = le.inverse_transform(predictions)
actual_labels = le.inverse_transform(y_test)
```

```python
# Evaluate
accuracy = accuracy_score(y_test, predictions)
```

```python
print("âœ… Accuracy:", accuracy)
print("ðŸ§  Predicted:", decoded_predictions)
print("ðŸŽ¯ Actual:   ", actual_labels)
```

```python
# Optional: Show neighbors of the first test sample
distances, indices = knn.kneighbors([X_test[0]])
print("\nðŸ”Ž Nearest Neighbors of 1st Test Sample:")
for i in range(len(indices[0])):
    label = le.inverse_transform([y_train[indices[0][i]]])[0]
    print(f"{i+1}. Label: {label} | Distance: {distances[0][i]:.2f}")
```
