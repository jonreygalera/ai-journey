```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

## Load the Data

```python
data = {'Height': [160, 170, 180, 150, 165, 155, 175, 158, 172, 153],
        'Weight': [55, 70, 85, 45, 65, 50, 80, 52, 77, 49],
        'Type': ['Burger', 'Pizza', 'Pizza', 'Burger', 'Pizza', 'Burger', 'Pizza', 'Burger', 'Pizza', 'Burger']}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=1)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
predictions = knn.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
report = classification_report(y_test, predictions, target_names=le.classes_)
```

```python
print(predictions)
print(f'Accuracy: {accuracy}')
print(report)
```
